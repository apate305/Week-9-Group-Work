# A. The Difference Between Linear and Rank Correlation

Linear correlation is the relationship between two variables that have a constant rate of change. 

In rank correlation both variables do change together BUT not at a constant rate. 

# B. Scatter Plot: Relationship between calculated_host_listings_count & availability_365

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#modules for statistics
import math
import statistics
from scipy import stats

belize_listings_path = "belize_airbnb_listings copy 2.csv"
belize_df = pd.read_csv(belize_listings_path)
belize_df.head()
```

```python
fig, ax = plt.subplots()

ax.plot(belize_df['calculated_host_listings_count'], belize_df['availability_365'], marker = "o", linestyle='None', color= 'blue')

# Set the x-axis label
ax.set_xlabel('Calculated Host Listings Count')

# Set the y-axis label 
ax.set_ylabel('Availability_365')

# Set the title
ax.set_title("Relationship between Calculated Host Listings Count & Availability 365")

plt.grid()


plt.show()
```

2. Create a box and whisker plot for the Belize Airbnb data price. In a markdown cell, describe what is going on in as simple of terms as possible. Use the documentation https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.boxplot.html



```python
# Creating plot
fig, ax = plt.subplots(1,1, figsize=(40,40))
plt.boxplot(belize_df['price'])

ax.tick_params(axis='both', which='major', labelsize=25)

ax.set_title("Belize Airbnb Data Price", fontsize= 60)

# show plot
plt.show()

```

This graph allows us to see the spread of the price data. The box contains all the prices that are close to the median price or the middle half of the data. The botttom whisker displays the range of prices below the middle half and the top whisker displays the prices that are above the middle half. The points above the top whisker are the outliers. The four biggest outliers are what stretch out the data.
